<h3 style="text-align:center; color:#4FFFB9;">Array Traversal</h3>
<h4 style="text-align:center; color:#FF69B4;">*****Remove duplicates*****</h4>
<p style="text-align:center; color:#538FED;">Given an array. We have to find duplicates and collect all distinctive values in another array. Finally print the new array.</p>
<div style="text-align: center;">
  <ol style="display: inline-block; text-align: left; color:#53DEED;">
     <li>Hash-Set Approach</li>
     <li>Normal Approach</li>
  </ol>
</div>
<p style="text-align:center; color:#D353ED;">Both approach works for O(n) time complexity.</p>

```python
#Hash-Set or dictionary approach
arr = [1 , 2 , 2 , 3 , 4 , 4 , 4 , 5 , 5]
#Function to remove duplicates from the list
def removeduplicate(arr):
    tr = set() #This is a set used to keep track of unique elements seen so far
    index = 0 #Index keeps track of the position where the next unique element will be placed
    #Loop through every element in the list
    for i in range(len(arr)):
        if arr[i] not in tr: #If current element has not been seen before
            tr.add(arr[i]) #Add it to the set of seen elements
            arr[index] = arr[i] #Place this unique element at the current "index" position
            index += 1 #Move index forward for the next unique element. It interprets as index = index + 1
    return index #Return the number of unique elements (new length of unique part of array)
#Main program execution
if __name__ == "__main__":
    new_arr = removeduplicate(arr) #Calling the functiona and storing its response in a new variable
    #Print only the unique elements (from index 0 to new_arr)
    for i in range(new_arr):
        print(arr[i], end=" ") #print unique elements in one line seperated by space
```

```python
#Normal approach
arr = [1 , 2 , 2 , 3 , 4 , 4 , 4 , 5 , 5]
#Function to remove duplicate from a sorted array
def removeduplicate(arr):
    n = len(arr)
    if n <= 1 : #If array has 0 or 1 element, no duplicate exist
        return n
    #The first element is always unique, so start from 1
    index = 1 #Initialize index for placing next unique element
    #Lop through array starting from second element
    for i in range(1 , n):
        if arr[i] != arr[i - 1]: #Compare current element with previous element
            arr[index] = arr[i] #Place it at the current "index"
            index += 1 #increament index for next unique element
    return index #Return the number of unique elements
#Main block to run the execution:
if __name__ == "__main__":
    new_arr = removeduplicate(arr) #Calling the function
    for i in range(new_arr): #Print only unique elements
        print(arr[i], end=" ") #Print elements in a single line seperated by a space
```

<h3 style="text-align:center; color:#4FFFB9;">Array Traversal</h3>
<h4 style="text-align:center; color:#FF69B4;">*****Check if an array is sorted*****</h4>
<p style="text-align:center; color:#538FED;">Given a array. We have to find if that array is sorted or not. If sorted it will print true otherwise false.</p>
<div style="text-align: center;">
  <ol style="display: inline-block; text-align: left; color:#53DEED;">
     <li>Iterative Approach</li>
     <li>Recursive Approach</li>
      <li>Using built-in function Approach</li>
  </ol>
</div>
<p style="text-align:center; color:#D353ED;">All approach works for O(n) time complexity.</p>


```python
#Iterative approach
arr = [10 , 20 , 30 , 40 , 50 , 60]
#Declaring a function which will check every element inside that array
def is_that_sorted(arr):
    n = len(arr) #Get the length of thta given array
    #Loop through the array starting from the second element
    for i in range(1 , n):
        if arr[i - 1] > arr[i]: #Compare current element with the previous element
            return False #If previous element is greater array is not sorted
    return True #If loop completes without returning false then array shoul be sorted and it will return True
#Main block of code to run the function
if __name__ == "__main__":
    print(is_that_sorted(arr)) #call the function and print the response (True if sorted, False is not)
```

```python
#Recursive Approach
arr = [10 , 20 , 30 , 40 , 50 , 60]
#In this approach we will use two function
def sort_helper(arr , n): 
    if n == 0 or n == 1: #Base case
        return True #Return true because an empty or single element array doesnt need to check
    #check if the last element is >= the second last element
    return (arr[n - 1]) >= arr[n - 2] and sort_helper(arr, n-1) #AND recursively check the rest of the elements
#Driver function
def is_this_sorted(arr):
    n = len(arr) #get the length of the array
    return sort_helper(arr , n) #call the function to check recursively
#Main block
if __name__ == "__main__":
    print(is_this_sorted(arr)) #Call the driver function to check the array. It will return True if sorted and False if unsorted
```

```python
#Built in function approach
arr = [10 , 20 , 30 , 40 , 50 , 60]
def builtin_sorted(arr):
    #Compare it with the original array to see if the original array ar equal to the sorted array
    return arr == sorted(arr) #sorted(arr) return a new list with elements in ascending order
    #arr == sorted(arr) checks element by element if the original array is identical to new list
#main block
if __name__ == "__main__":
    print(builtin_sorted(arr)) #Call the function and print the result (True if sorted, False if not)
```
